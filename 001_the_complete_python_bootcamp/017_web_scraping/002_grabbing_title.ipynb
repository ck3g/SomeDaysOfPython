# Grabbing a page title

```python
import requests
```

```python
result = requests.get('https://de.wikipedia.org/wiki/Berlin')
```

```python
type(result)
```

```python
result.text
```

```python
import bs4
```

```python
soup = bs4.BeautifulSoup(result.text, 'lxml')
```

```python
soup
```

```python
soup.select('title')
```

```python
soup.select('title')[0].getText()
```
